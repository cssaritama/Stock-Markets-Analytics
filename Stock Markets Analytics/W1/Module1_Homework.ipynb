```python
# Import necessary libraries
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Download S&P 500 data from Wikipedia
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies "
tables = pd.read_html(url)
sp500_data = tables[0]  # First table contains the relevant data
```

```python
# Process the 'Date added' column
sp500_data['Date added'] = pd.to_datetime(sp500_data['Date added'], errors='coerce')
sp500_data['Year added'] = sp500_data['Date added'].dt.year

# Count additions per year (exclude 1957)
additions_per_year = sp500_data['Year added'].value_counts().sort_index()
additions_per_year = additions_per_year[additions_per_year.index > 1957]

# Find the year with the most additions
year_with_most_additions = additions_per_year.idxmax()
print(f"The year with the most additions is: {year_with_most_additions}")
```

```python
# Save results to a CSV file
additions_per_year.to_csv("sp500_additions_per_year.csv")
```

```python
import pandas as pd
import yfinance as yf

# List of index symbols
symbols = [
    "^GSPC", "000001.SS", "^HSI", "^AXJO", "^NSEI",
    "^GSPTSE", "^GDAXI", "^FTSE", "^N225", "^MXX", "^BVSP"
]

# Download historical data for indices
start_date = "2025-01-01"
end_date = "2025-05-01"
data = yf.download(symbols, start=start_date, end=end_date)

# Check the structure of the data
print("Data structure:")
print(data.head())

# Select only the 'Close' prices
close_data = data.xs('Close', level='Price', axis=1)

# Calculate YTD returns
ytd_returns = (close_data.iloc[-1] / close_data.iloc[0]) - 1

# Compare with S&P 500
sp500_return = ytd_returns["^GSPC"]
better_than_sp500 = (ytd_returns > sp500_return).sum()

print(f"Number of indices with better YTD returns than S&P 500: {better_than_sp500}")

# Save results to a CSV file
ytd_returns.to_csv("index_ytd_returns.csv")
```

```python
import pandas as pd
import yfinance as yf

# List of index symbols
symbols = [
    "^GSPC", "000001.SS", "^HSI", "^AXJO", "^NSEI",
    "^GSPTSE", "^GDAXI", "^FTSE", "^N225", "^MXX", "^BVSP"
]

# Download historical data for indices
start_date = "2025-01-01"
end_date = "2025-05-01"
data = yf.download(symbols, start=start_date, end=end_date)

# Check the structure of the data
print("Data structure:")
print(data.head())

# Select only the 'Close' prices
close_data = data.xs('Close', level='Price', axis=1)

# Calculate YTD returns
ytd_returns = (close_data.iloc[-1] / close_data.iloc[0]) - 1

# Compare with S&P 500
sp500_return = ytd_returns["^GSPC"]
better_than_sp500 = (ytd_returns > sp500_return).sum()

print(f"Number of indices with better YTD returns than S&P 500: {better_than_sp500}")

# Save results to a CSV file
ytd_returns.to_csv("index_ytd_returns.csv")
```

```python
# Download historical data for S&P 500
sp500_data = yf.download("^GSPC", start="1950-01-01")['Close']

# Identify all-time highs
all_time_highs = sp500_data[sp500_data == sp500_data.cummax()]

# Calculate corrections
corrections = []
for i in range(1, len(all_time_highs)):
    high_prev = all_time_highs.iloc[i - 1]
    high_curr = all_time_highs.iloc[i]

    # Ensure min_price is calculated correctly
    min_price = sp500_data.loc[high_prev.name:high_curr.name].min()
    drawdown = (high_prev - min_price) / high_prev * 100  # This should be a scalar

    # Debugging: Print values to verify
    print(f"high_prev: {high_prev}, high_curr: {high_curr}, min_price: {min_price}, drawdown: {drawdown}")

    # Check if drawdown is a scalar and significant
    if isinstance(drawdown, (int, float)) and drawdown >= 5:
        duration = (high_curr.name - high_prev.name).days
        corrections.append(duration)

# Debugging: Print the number of corrections found
print(f"Number of corrections detected: {len(corrections)}")

# Calculate percentiles only if corrections are found
if corrections:
    import numpy as np
    percentiles = np.percentile(corrections, [25, 50, 75])
    median_duration = percentiles[1]
    print(f"Median correction duration: {median_duration} days")
else:
    print("No significant corrections detected.")

# Save results to a CSV file
pd.DataFrame({"Correction Durations": corrections}).to_csv("sp500_correction_durations.csv")
```
